```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
manhattan = pd.read_csv('manhattan.csv')
manhattan.head()
```

```python
inversion = pd.read_csv('inversion.csv')
inversion.head()
```

```python
x_lim = (
    min(manhattan.path_length.min(), manhattan.path_length.min()) - 1,
    max(manhattan.path_length.max(), inversion.path_length.max()) + 1
)
x_range = range(*x_lim)
```

```python
plt.title('Solving time [s]')
plt.xlabel('Optimal solution path length')
plt.scatter(manhattan.path_length, manhattan.solving_time,
            color='blue', label='manhattan distance')
plt.scatter(inversion.path_length, inversion.solving_time,
            color='red', label='inversion distance')
plt.plot(x_range, [manhattan.solving_time.mean()] * len(x_range),
         color='blue', linewidth=2.5)
plt.plot(x_range, [inversion.solving_time.mean()] * len(x_range),
         color='red', linewidth=2.5)
plt.xlim(*x_lim)
plt.legend()
plt.savefig('time.png')
plt.show();
```

```python
plt.title('Number of visited nodes')
plt.xlabel('Optimal solution path length')
plt.scatter(manhattan.path_length, manhattan.visited_nodes,
            color='blue', label='manhattan distance')
plt.scatter(inversion.path_length, inversion.visited_nodes,
            color='red', label='inversion distance')
plt.plot(x_range, [manhattan.visited_nodes.mean()] * len(x_range),
         color='blue', linewidth=2.5)
plt.plot(x_range, [inversion.visited_nodes.mean()] * len(x_range),
         color='red', linewidth=2.5)
plt.xlim(*x_lim)
plt.legend()
plt.savefig('visited.png')
plt.show();
```
